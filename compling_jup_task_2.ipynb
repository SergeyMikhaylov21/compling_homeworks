```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

from gensim.summarization import keywords
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import stopwords
```

```python
words, clean_words, stemmed_words = [], [], [] 
stopWords = set(stopwords.words('english')) #используем стоп-слова из пакета nltk 

with open('data_for_keyword.txt', encoding = 'utf-8') as corpusfile: #текстовый файл - тексты нескольких научных статей
    lines = corpusfile.read().splitlines()
    for line in lines:
        wrd = line.split(' ')
        for elem in range(len(wrd)): #очищаем данные
            x = wrd[elem].strip().replace(',', '').replace(' ', '').replace('-', '').replace('(', '')
            y = x.replace(')', '').replace('.', '').replace('[', '').replace(']', '').lower()
            words.append(y)
```

```python
for elem in words: #убираем стоп-слова
    if elem not in stopWords:
        clean_words.append(elem)
    else:
        continue
```

```python
stemmer = SnowballStemmer("english") #стемминг
for wrd in clean_words:
    wrd1 = stemmer.stem(str(wrd))
    stemmed_words.append(wrd1)
print(stemmed_words)
```

```python
stemmed_words = str(stemmed_words) #извлекаем ключевые слова
print(keywords(stemmed_words, ratio=0.01))
```

Описание использованного способа поика ключевых слов можно найти по ссылке: https://rare-technologies.com/text-summarization-with-gensim/

Если вместо стемминга проводить лемматизацию, результаты получаются хуже. Но в общем и целом, для данного задания необходим предварительный pos-теггинг, который значительно улучшит качество результатов (как в случае предобработки со стеммингом, так и в случае с лемматизацией).
